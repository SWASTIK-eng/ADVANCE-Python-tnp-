```python
balance = 0.0
transactions = []

pin = input("Create a 4-digit PIN: ")

attempts = 3

def verify_pin():
    global attempts
    entered = input("Enter your PIN: ")

    if entered == pin:
        attempts = 3
        return True
    else:
        attempts -= 1
        print(f"Wrong PIN! Attempts left: {attempts}")
        if attempts == 0:
            print("Account locked. Too many wrong attempts.")
            exit() # Exit the program if attempts run out
        return False

def deposit_money():
    global balance
    global transactions
    amount = float(input("Enter amount to deposit: "))
    if amount > 0:
        balance += amount
        transactions.append(f"Deposited: ${amount:.2f}")
        print(f"${amount:.2f} deposited successfully.")
    else:
        print("Deposit amount must be positive.")

def withdraw_money():
    global balance
    global transactions
    amount = float(input("Enter amount to withdraw: "))
    if amount <= 0:
        print("Withdrawal amount must be positive.")
    elif amount > balance:
        print("Insufficient funds.")
    else:
        balance -= amount
        transactions.append(f"Withdrew: ${amount:.2f}")
        print(f"${amount:.2f} withdrawn successfully.")

def check_balance():
    print(f"Current Balance: ${balance:.2f}")

def view_transaction_history():
    print("Transaction History:")
    if not transactions:
        print("No transactions yet.")
    for transaction in transactions:
        print(transaction)


while True:
    print("\nWelcome to the Banking System")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. View Transaction History")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    if choice == '5':
        print("Thank you for using the Banking System. Goodbye!")
        break # Exit the loop
    elif choice in ['1', '2', '3', '4']:
        if not verify_pin():
            continue # If PIN is wrong, re-display the menu

        # If PIN is correct, perform the chosen operation
        if choice == '1':
            deposit_money()
        elif choice == '2':
            withdraw_money()
        elif choice == '3':
            check_balance()
        elif choice == '4':
            view_transaction_history()
    else:
        print("Invalid choice. Please try again.")
```

```python
def count_vowels_consonants(text):
    vowels = "aeiouAEIOU"
    v_count = 0
    c_count = 0

    for ch in text:
        if ch.isalpha():
            if ch in vowels:
                v_count += 1
            else:
                c_count += 1

    return v_count, c_count


string = input("Enter a string: ")

vowels, consonants = count_vowels_consonants(string)

print("Number of vowels:", vowels)
print("Number of consonants:", consonants)
```

```python
# LIBRARY MANAGEMENT SYSTEM
def add_book(library):
    book_name = input("Enter book name: ")

    if book_name in library:
        print("Book already exists!\n")
        return

    author = input("Enter author name: ")

    library[book_name] = {
        "author": author,
        "available": True
    }

    print("Book added successfully!\n")



def borrow_book(library, borrowed_books):
    book_name = input("Enter book name to borrow: ")

    if book_name in library and library[book_name]["available"]:
        student_name = input("Enter student name: ")
        borrow_day = int(input("Enter borrowing day (number): "))

        due_day = borrow_day + 7

        library[book_name]["available"] = False

        borrowed_books[book_name] = {
            "student": student_name,
            "borrow_day": borrow_day,
            "due_day": due_day
        }

        print("Book borrowed successfully!")
        print("Due date (Day number):", due_day, "\n")
    else:
        print("Book not available!\n")

def return_book(library, borrowed_books):
    book_name = input("Enter book name to return: ")

    if book_name in borrowed_books:
        return_day = int(input("Enter return day (number): "))
        due_day = borrowed_books[book_name]["due_day"]

        late_days = return_day - due_day

        if late_days > 0:
            fine = late_days * 10
            print("Late fine is: â‚¹", fine)
        else:
            print("No fine!")

        library[book_name]["available"] = True
        del borrowed_books[book_name]

        print("Book returned successfully!\n")
    else:
        print("This book was not borrowed!\n")



def display_books(library, borrowed_books):
    print("\nLibrary Books:")

    for book in library:
        author = library[book]["author"]
        status = "Available"
        user = "-"
        due = "-"

        if book in borrowed_books:
            status = "Borrowed"
            user = borrowed_books[book]["student"]
            due = borrowed_books[book]["due_day"]

        print("Title:", book)
        print("Author:", author)
        print("Status:", status)
        print("Borrowed By:", user)
        print("Due Day:", due)
        print("-------------------------")

    print("Total Books:", len(library), "\n")



def main():
    library = {
        "Python Programming": {"author": "Guido van Rossum", "available": True},
        "Data Structures": {"author": "Mark Allen Weiss", "available": True},
        "Algorithms": {"author": "Thomas H. Cormen", "available": True},
        "Operating System Concepts": {"author": "Abraham Silberschatz", "available": True},
        "Computer Networks": {"author": "Andrew S. Tanenbaum", "available": True},
        "Database System Concepts": {"author": "Henry F. Korth", "available": True},
        "Artificial Intelligence": {"author": "Stuart Russell", "available": True},
        "Machine Learning": {"author": "Tom M. Mitchell", "available": True},
        "Clean Code": {"author": "Robert C. Martin", "available": True},
        "The Pragmatic Programmer": {"author": "Andrew Hunt", "available": True}
    }

    borrowed_books = {}

    while True:
        print("1. Add Book")
        print("2. Borrow Book")
        print("3. Return Book")
        print("4. Display Books")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_book(library)
        elif choice == "2":
            borrow_book(library, borrowed_books)
        elif choice == "3":
            return_book(library, borrowed_books)
        elif choice == "4":
            display_books(library, borrowed_books)
        elif choice == "5":
            print("Exiting Program...")
            break
        else:
            print("Invalid choice!\n")


main()
```

```python
# STUDENT GRADEBOOK
def get_grade(marks):
    if 90 <= marks <= 100:
        return "O"
    elif 80 <= marks < 90:
        return "E"
    elif 70 <= marks < 80:
        return "A"
    elif 60 <= marks < 70:
        return "B"
    elif 50 <= marks < 60:
        return "C"
    elif 40 <= marks < 50:
        return "D"
    else:
        return "F"

def student_gradebook_system():
    gradebook = {}

    while True:
        print(" STUDENT GRADEBOOK SYSTEM ")
        print("1. Add Student Grade")
        print("2. Display Gradebook")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            student_name = input("Enter student name: ").strip()
            if not student_name:
                print("Student name cannot be empty.")
                continue

            while True:
                try:
                    marks = float(input(f"Enter marks for {student_name} (0-100): "))
                    if 0 <= marks <= 100:
                        break
                    else:
                        print("Marks must be between 0 and 100.")
                except ValueError:
                    print("Invalid input. Please enter a number for marks.")

            grade = get_grade(marks)
            gradebook[student_name] = {'marks': marks, 'grade': grade}
            print(f"Grade for {student_name} (Marks: {marks}) added as {grade}.")

        elif choice == '2':
            if not gradebook:
                print("Gradebook is empty.")
            else:
                print("\n--- CURRENT GRADEBOOK ---")
                for name, data in gradebook.items():
                    print(f"Student: {name}, Marks: {data['marks']:.2f}, Grade: {data['grade']}")
                print("-------------------------")

        elif choice == '3':
            print("Exiting Gradebook System. Goodbye!")
            break:

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

student_gradebook_system()
```

```python
# TRAFFIC LIGHT SIMULATION
import time

def red():
    print(" RED - Stop")
    time.sleep(2)


def yellow():
    print(" YELLOW - Wait")
    time.sleep(1)


def green():
    print(" GREEN - Go")
    time.sleep(2)


def traffic():
    cycles = int(input("Enter number of signal cycles: "))

    for i in range(cycles):
        print("\nCycle", i+1)
        red()
        yellow()
        green()


traffic()
```

```python
# ONLINE SHOPPING CART

products = {
    "apple": 1.0,
    "banana": 0.5,
    "orange": 0.75
}

def add_to_cart(cart, item_name, quantity=1):
    if item_name.lower() in products:
        if item_name.lower() in cart:
            cart[item_name.lower()] += quantity
        else:
            cart[item_name.lower()] = quantity
        print(f"{quantity} x {item_name} added to cart.")
    else:
        print(f"Sorry, {item_name} is not available.")

def view_cart(cart):
    if not cart:
        print("Your cart is empty.")
        return

    print("\n--- Your Shopping Cart ---")
    total_price = 0
    for item, quantity in cart.items():
        price = products[item] * quantity
        total_price += price
        print(f"{item.capitalize()} (x{quantity}) - ${products[item]:.2f} each = ${price:.2f}")
    print(f"-------------------------")
    print(f"Total: ${total_price:.2f}")
    print("-------------------------\n")

def simple_shopping_cart():
    cart = {}
    while True:
        print("\n--- SIMPLE SHOPPING CART MENU ---")
        print("1. Add Item")
        print("2. View Cart")
        print("3. Exit")
        print("-----------------------------------")

        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            item = input("Enter item name to add: ").lower()
            try:
                qty = int(input(f"Enter quantity for {item} (default is 1): ") or 1)
                if qty < 1: raise ValueError
            except ValueError:
                print("Invalid quantity. Please enter a positive number.")
                continue
            add_to_cart(cart, item, qty)
        elif choice == '2':
            view_cart(cart)
        elif choice == '3':
            print("Exiting simple shopping cart. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 3.")

simple_shopping_cart()
```
